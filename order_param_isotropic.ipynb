```python
import numpy as np 
import pandas as pd
```

```python
col_names=['px','py', 'pz']
df = pd.read_excel('data_isotropic_gel.xlsx', names=col_names, header= None)

df
```

```python
df['px2'] = df['px']**2
df['py2'] = df['py']**2
df['pz2'] = df['pz']**2
df['pxpy'] = df['px']*df['py']
df['pxpz'] = df['px']*df['pz']
df['pypz'] = df['py']*df['pz']
df
```

```python
S_vector = np.zeros((3,3))

S_vector[0,0] = df['px2'].mean()
S_vector[1,1] = df['py2'].mean()
S_vector[2,2] = df['pz2'].mean()
S_vector[0,1] = df['pxpy'].mean()
S_vector[1,0] = S_vector[0,1]
S_vector[0,2] = df['pxpz'].mean()
S_vector[2,0] = S_vector[0,2]
S_vector[1,2] = df['pypz'].mean()
S_vector[2,1] = S_vector[1,2]
S_vector = 1.5*S_vector -0.5 *np.eye (3)

S_vector
```

```python
eigenvalues, eigenvectors = np.linalg.eig(S_vector)

# Find index of largest eigenvalue
max_index = np.argmax(eigenvalues)

# Scalar order parameter
S = eigenvalues[max_index]

# Director vector (unit vector)
n = eigenvectors[:, max_index]

print("Scalar order parameter S =", S)
print("Director vector n =", n)
```
